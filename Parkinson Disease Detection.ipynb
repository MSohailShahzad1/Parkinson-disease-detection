```python
from imblearn.over_sampling import SMOTE

# Separate features and target variable
X = data.drop('status', axis=1)
y = data['status']

# Apply SMOTE to oversample the minority class
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Create a new balanced DataFrame
balanced_data = pd.DataFrame(X_resampled, columns=X.columns)
balanced_data['status'] = y_resampled
data = balanced_data
# Visualize the balanced target variable
sns.countplot(x='status', data = data, palette='viridis', hue='status')
plt.title('Balanced Target Feature Distribution')
plt.show()
```

# **Feature Engeerinig**

Separate Features and Target

```python
X = data.drop(columns=['status'])
y = data['status']
```

Scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaledx = scaler.fit_transform(X)
```

Selecting important features

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95)
pcaX = pca.fit_transform(scaledx)
pcaX_df = pd.DataFrame(pcaX)
```

# **Data splitting for training and testing**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(pcaX_df, y, test_size=0.2, random_state=11)
# print the shape of train and test data
print("X_train shape: ", X_train.shape)
print("y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test shape: ", y_test.shape)
```

# **k-Nearest Neighbors (k-NN)**

Importing Library and creating KNN object

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

Training the model

```python
knn.fit(X_train, y_train)
```

Making Prediction on testing data

```python
y_pred = knn.predict(X_test)
y_pred[0:10]
```

As this is a Classification type algorithm. So, using confusion matrix to display it

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred), display_labels=["Not Parkinson", "Parkinson"])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

Now checking the effeciency of model using performance matrices

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)


print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```

# **Ensemble Learning**

importing necessary library

```python
from sklearn.ensemble import VotingClassifier
```

creating individual classifier

```python
knn_classifier = KNeighborsClassifier(n_neighbors=3)
lr_classifier = LogisticRegression()
nb_classifier = GaussianNB()
dt_classifier = DecisionTreeClassifier(random_state=42)
```

creating a voting classifier Voting **HARD**

```python
estimators = [('knn', knn_classifier), ('lr', lr_classifier), ('nb', nb_classifier), ('dt', dt_classifier)]
ensemble_classifier = VotingClassifier(estimators=estimators, voting='hard')
```

Training the model

```python
ensemble_classifier.fit(X_train, y_train)
```

```python
y_pred = ensemble_classifier.predict(X_test)
```

# **Ensemble Learning Stacking**

```python
from sklearn.ensemble import StackingClassifier
```

```python
base_learners = [
    ('knn', KNeighborsClassifier(n_neighbors=3)),
    ('lr', LogisticRegression()),
    ('nb', GaussianNB()),
    ('dt', DecisionTreeClassifier(random_state=42)),
]

meta_learner = LogisticRegression()

stacking_classifier = StackingClassifier(
    estimators=base_learners,
    final_estimator=meta_learner,
    cv=5,
    stack_method='predict_proba',
)
```

```python
stacking_classifier.fit(X_train, y_train)
```

```python
y_pred_stacking = stacking_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred_stacking)
precision = precision_score(y_test, y_pred_stacking)
recall = recall_score(y_test, y_pred_stacking)
f1 = f1_score(y_test, y_pred_stacking)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```
